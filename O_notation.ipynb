```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a range of values for n
n = np.linspace(1, 100, 500)
n_log_n = n * np.log(n)
n_squared = n ** 2

# Plot each complexity
plt.figure(figsize=(10, 6))
plt.plot(n, np.ones_like(n), label='O(1)')
plt.plot(n, np.log(n), label='O(log n)')
plt.plot(n, n, label='O(n)')
plt.plot(n, n_log_n, label='O(n log n)')
plt.plot(n, n_squared, label='O(n^2)')

# Adding labels and title
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.title('Comparison of Time Complexities')
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating a range of values for n (I did not include 1 to avoid errors)
n = np.linspace(1, 100, 500)

# Calculating different order
O1 = np.ones_like(n)                  # O(1)
O_log_n = np.log(n)                   # O(log n)
O_n = n                               # O(n)
O_n_log_n = n * np.log(n)             # O(n log n)
O_n_squared = np.power(n, 2)          # O(n^2)
O_2_power_n = np.power(2, n)          # O(2^n)
O_n_factorial = [np.math.factorial(i) for i in range(1, len(n) + 1)]  # O(n!)

# charts for the complexities of order
plt.figure(figsize=(10, 6))
plt.plot(n, O1, label='O(1)')
plt.plot(n, O_log_n, label='O(log n)')
plt.plot(n, O_n, label='O(n)')
plt.plot(n, O_n_log_n, label='O(n log n)')
plt.plot(n, O_n_squared, label='O(n^2)')
plt.plot(n, O_2_power_n, label='O(2^n)', color='purple')
plt.plot(n[:20], O_n_factorial[:20], label='O(n!)', color='black')  # Limited to n=20 for visibility

# Adding labels and title
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.yscale('log')  # Using logarithmic scale due to wide range of values
plt.title('Comparison of Various Time Complexities')
plt.legend()

# Display the chart
plt.show()
```

```python

import numpy as np
import time

# O(1) - Constant Time Complexity
def constant_time_example():
    array = [1, 2, 3, 4, 5]
    first_element = array[0]
    print("First element of the array:", first_element)

# O(log n) - Logarithmic Time Complexity
def logarithmic_time_example():
    sorted_array = sorted([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
    index = binary_search(sorted_array, 30)
    print("Index of element 30:", index)

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# O(n) - Linear Time Complexity
def linear_time_example():
    array = [1, 2, 3, 4, 5]
    array_sum = sum(array)
    print("Sum of array elements:", array_sum)

# O(n log n) - Linearithmic Time Complexity
def linearithmic_time_example():
    array = [5, 2, 9, 1, 6, 3]
    sorted_array = merge_sort(array)
    print("Sorted array:", sorted_array)

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# O(n^2) - Quadratic Time Complexity
def quadratic_time_example():
    array = [5, 2, 9, 1, 6, 3]
    selection_sorted_array = selection_sort(array)
    print("Array sorted using selection sort:", selection_sorted_array)

def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# O(2^n) - Exponential Time Complexity
def exponential_time_example():
    n = 10
    subsets = generate_subsets(n)
    print("All subsets of a set of size", n, ":", subsets)

def generate_subsets(n):
    subsets = []
    for i in range(2**n):
        subset = []
        for j in range(n):
            if (i >> j) & 1:
                subset.append(j)
        subsets.append(subset)
    return subsets

# O(n!) - Factorial Time Complexity
def factorial_time_example():
    n = 3
    permutations = generate_permutations(n)
    print("All permutations of", n, "elements:", permutations)

def generate_permutations(n):
    from itertools import permutations
    return list(permutations(range(n)))

if __name__ == "__main__":
    print("Examples of Different Time Complexities:")
    print("--------------------------------------")

    print("\nO(1) - Constant Time Complexity:")
    start_time = time.time()
    constant_time_example()
    print("Execution Time:", time.time() - start_time)

    print("\nO(log n) - Logarithmic Time Complexity:")
    start_time = time.time()
    logarithmic_time_example()
    print("Execution Time:", time.time() - start_time)

    print("\nO(n) - Linear Time Complexity:")
    start_time = time.time()
    linear_time_example()
    print("Execution Time:", time.time() - start_time)

    print("\nO(n log n) - Linearithmic Time Complexity:")
    start_time = time.time()
    linearithmic_time_example()
    print("Execution Time:", time.time() - start_time)

    print("\nO(n^2) - Quadratic Time Complexity:")
    start_time = time.time()
    quadratic_time_example()
    print("Execution Time:", time.time() - start_time)

    print("\nO(2^n) - Exponential Time Complexity:")
    start_time = time.time()
    exponential_time_example()
    print("Execution Time:", time.time() - start_time)

    print("\nO(n!) - Factorial Time Complexity:")
    start_time = time.time()
    factorial_time_example()
    print("Execution Time:", time.time() - start_time)
```
